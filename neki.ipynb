```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from collections import Counter
```

```python
VIDEO_NUM = 2
root_image_data = f'data/img/{VIDEO_NUM}'
green = []
red = []
blue = []
images = []
for img in sorted(
    os.listdir(root_image_data), 
                  key=lambda x: int(x.split('frame')[1][:-4])
                  ):
    img_array = cv2.imread(os.path.join(root_image_data, img))
    print(os.path.join(root_image_data, img))

    green.append(
        np.sum(img_array[:, :, 1])
    )
    blue.append(
        np.sum(img_array[:, :, 0])
    )
    red.append(
        np.sum(img_array[:, :, 2])
    )
    images.append(img_array)

    # plt.imshow(img_array, cmap='gray')
    # plt.show()
    # break
```

```python
start = red.index(max(red[:250]))
end = red.index(max(red[250:]))
print(start, end)
```

```python
loc = []
c = 0
cur = start + 5
mode = "green"
while cur < end:
    if mode == "green":
        cur += green[cur:(cur + 4)].index(max(green[cur:(cur + 4)]))
        mode = "red"
    else:
        cur += green[cur:(cur + 4)].index(min(green[cur:(cur + 4)]))
        loc.append([c, cur])
        c += 1
        mode = "green"
```

```python
r = []
for i in loc:
    if np.max(images) > 200:
        x = np.where(images == np.max(images))
        r.append([i[0], x])
```
